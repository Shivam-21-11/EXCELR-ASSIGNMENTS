# Model Building

```python
df.rename(columns={'R&D Spend':'RD_Spend',
                  'Marketing Spend':'Marketing_Spend'},inplace=True)
```

```python
model = sfa.ols("Profit~RD_Spend+Administration+Marketing_Spend+State",data=df).fit()
```

```python
model.rsquared
```

```python
model.pvalues
```

# VIF

```python
vif_stat = pd.DataFrame()
vif_stat['Feature'] = df.columns[:-1]
vif_stat['vif'] = [variance_inflation_factor(df.values,i) for i in range(len(df.columns[:-1]))]
vif_stat.head()
```

- *Administration column having high variance_inflation_factor*

# Subset Collection

```python
#RD_spend model
rd_model = sfa.ols('Profit~RD_Spend+Marketing_Spend+State',data=df).fit()
```

```python
rd_model.aic,rd_model.rsquared
```

```python
#Administration model
ad_model = sfa.ols('Profit~Administration+Marketing_Spend+State',data=df).fit()
```

# Residual Analysis

```python
fmodel = sfa.ols('Profit~RD_Spend+Marketing_Spend+State',data=df).fit()
```

```python
sa.qqplot(fmodel.resid,line='q');
```

# model deletion technique

# cooks distance

```python
inf = fmodel.get_influence()
c,p = inf.cooks_distance
```

```python
plt.stem(c)
```

**No value exceed the limit value of 1 for cooks distance**

```python
# Calculate Cutoff
k = len(df.columns)
n = len(df)

levarge_cutoff = 3*(k+1)/n
levarge_cutoff
```

```python
influence_plot(model)
plt.axvline(levarge_cutoff,color = 'red',linestyle = '--')
```

# sqrt transformation

```python
sq_df = pd.DataFrame()
for i,col in enumerate(df.columns):
    if i < len(df.columns) - 2:
        sq_df[f'log_{col}'] = np.sqrt(df[col])
    else:
        sq_df[col] = df[col]
sq_df.head()
```

```python
sq_model = sfa.ols('Profit~log_RD_Spend+log_Administration+log_Marketing_Spend+State',data=sq_df).fit()
sq_model.rsquared
```

# Final Model

```python
final_model = sfa.ols('Profit~RD_Spend+Marketing_Spend+State',data=df).fit()
```

```python
final_model.rsquared
```

```python
data = {
    'Model': ['Base Model','R&D Model', 'Administration Model','Sqrt Transformed Model','Final Model'],
    'R^2 Score': [model.rsquared,rd_model.rsquared,ad_model.rsquared,sq_model.rsquared,final_model.rsquared]
}
out_df = pd.DataFrame(data)
out_df
```
