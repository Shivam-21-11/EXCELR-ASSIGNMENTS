# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 250
import seaborn as sns
sns.set_theme(style='darkgrid',palette='rainbow')
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.cluster import hierarchy
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
from yellowbrick.cluster import SilhouetteVisualizer
```

# Data

```python
df = pd.read_csv('./wine.csv')
df = df.iloc[:,1:]
df
```

# Data PreProcessing

```python
df.describe().T
```

```python
sc = StandardScaler()
scaled = sc.fit_transform(df)
```

# PCA

```python
pca = PCA()
pd.DataFrame(pca.fit_transform(scaled))
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of features')
plt.ylabel('Explained Variance')
plt.title('Scree Plot', size = 20)
```

*Since problem statement specifically asked us to use first 3 components*

```python
pca = PCA(n_components=3)
pcs = pca.fit_transform(scaled)
pcs = pd.DataFrame(pcs, columns=['pc1','pc2','pc3'])
pcs
```

# Clustering

## 1. Hierarchical Clustering

```python
lk = hierarchy.linkage(pcs,method='single')
plt.axhline(2.9,linestyle='--',c='r')
dendo = hierarchy.dendrogram(lk)
```

*As observed in the dendogram the clustering is not optimal*

```python
lk_c = hierarchy.linkage(pcs,method='complete')
dendo = hierarchy.dendrogram(lk_c)
plt.axhline(7,linestyle='--',c='r')
plt.show()
```

*This gives ideal cluster n_cluster = 4*

```python
lk_avg = hierarchy.linkage(pcs,method='average')
dendo = hierarchy.dendrogram(lk_avg)
plt.axhline(4.2,linestyle='--',c='r')
plt.show()
```

```python
lk_ct = hierarchy.linkage(pcs,method='centroid')
dendo = hierarchy.dendrogram(lk_ct)
plt.axhline(4,linestyle='--',c='r')
plt.show()
```

**Ideal Number of Clusters as infered from the above methods is n_cluster = 4 using complete method**

```python
# on original data
lk_cg = hierarchy.linkage(scaled,method='complete')
dendo = hierarchy.dendrogram(lk_cg)
plt.axhline(8,linestyle='--',c='r')
plt.show()
```

*For original data we get 5 cluster but on pca data we get 4 cluster*

### Model

```python
# on pcs data
hc = AgglomerativeClustering(n_clusters=4)
y = hc.fit_predict(pcs)
```

```python
silhouette_score(pcs,y)
```

```python
pcs['Cluster(H)'] = y
```

```python
pcs.groupby('Cluster(H)').agg('mean',numeric_only=True)
```

```python
sns.scatterplot(x=pcs['pc1'],y= pcs['pc2'], hue = pcs['Cluster(H)'],palette='rainbow')
plt.show()
```

## 2. Kmeans

```python
km = KMeans()
y = km.fit_predict(pcs)
```

```python
km.cluster_centers_ # since default num_cluster = 8 we get 8 centers
```

```python
#Visualize the clusters
sns.scatterplot(x=pcs['pc1'],y=pcs['pc2'],hue = y, palette='rainbow')
```

```python
# evaluate 
silhouette_score(pcs,y)
```

### Elbow graph for optimal n_cluster

```python
# for pcs data
wcss = []

for i in range(1,20):
    km = KMeans(n_clusters=i)
    km.fit_predict(pcs)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,20),wcss)
plt.ylabel('WCSS')
plt.xlabel('N_Cluster')
plt.axvline(3,linestyle='--',c='r')
plt.show()
```

*Optimal number of cluster as infered from elbow graph is 3 for pcs data*

```python
# for original data
wcss_og = []

for i in range(1,20):
    km = KMeans(n_clusters=i)
    km.fit_predict(scaled)
    wcss_og.append(km.inertia_)
```

```python
plt.plot(range(1,20),wcss_og)
plt.ylabel('WCSS')
plt.xlabel('N_Cluster')
plt.axvline(3,linestyle='--',c='r')
plt.show()
```

*Optimal number of cluster as infered from elbow graph is 3 for original data same as we got for pcs data*

### Final model

```python
km = KMeans(n_clusters=3)
ypred = km.fit_predict(pcs)
ypred
```

```python
#Visualize the clusters
sns.scatterplot(x=pcs['pc1'],y=pcs['pc2'],hue = ypred, palette='rainbow')
```

```python
# Clusters with center
sns.scatterplot(x=pcs['pc1'],y=pcs['pc2'],hue = ypred, palette='rainbow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color = 'black',s=50)
```

*Cluster Center in black*

```python
#evaluation
silhouette_score(pcs,ypred)
```

```python
pcs['Cluster(kmean)'] = ypred
```

```python
pcs.groupby('Cluster(kmean)')[['pc1','pc2','pc3']].agg('mean',numeric_only=True)
```

### Kmean silhouette score visualization

```python
model = SilhouetteVisualizer(KMeans(n_clusters=3))
vis = model.fit(pcs)
vis.show()  
```
