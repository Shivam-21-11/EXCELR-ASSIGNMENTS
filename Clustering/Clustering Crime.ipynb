# Import

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from scipy.cluster import hierarchy
from sklearn.cluster import KMeans , DBSCAN
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer
from sklearn.neighbors import NearestNeighbors
import numpy as np
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (10,5)
import warnings
warnings.filterwarnings('ignore')
```

# Data

```python
df = pd.read_csv('./crime_data.csv',index_col=0)
df
```

# Viz

```python
sns.pairplot(df)
```

# Data processing

*We need to standardize the data since clustering algos perform distance calculation*

```python
std = StandardScaler()
scaled = std.fit_transform(df)
```

# 1. Hierarchical Clustering

**There are many linkage method we are going to try all and see which suits us better**

```python
lk = hierarchy.linkage(scaled,method='single') # single takes the shortest distance
dendo = hierarchy.dendrogram(lk)
plt.axhline(1.50,linestyle='--',c='r')
```

*As observed not ideal clustering*

```python
lk_c = hierarchy.linkage(scaled,method='complete') # single takes the farthest distance
dendo = hierarchy.dendrogram(lk_c)
plt.axhline(4,linestyle='--',c='r')
```

*As observed this gives us better cluster*

```python
lk_a = hierarchy.linkage(scaled,method='average') # single takes the average distance
dendo = hierarchy.dendrogram(lk_a)
plt.axhline(2.2,linestyle='--',c='r')
```

*As observed this also gives good cluster but still datapoint 1 is singled out*

```python
lk_ct = hierarchy.linkage(scaled,method='centroid')
dendo = hierarchy.dendrogram(lk_ct)
plt.axhline(1.7,linestyle='--',c='r')
```

**Clustering with complete method gives better clusters**

# Model

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=4)# here n_clusters is 4 since complete methods gives us 4 distinct clusters
y = hc.fit_predict(scaled)
```

```python
df['Cluster(H)'] = y
df.head()
```

```python
df.groupby('Cluster(H)').agg('mean')
```

## Some Viz

```python
sns.scatterplot(x=df['Murder'],y=df['Assault'],hue=df['Cluster(H)'],palette='rainbow')
```

```python
sns.scatterplot(x=df['Rape'],y=df['Assault'],hue=df['Cluster(H)'],palette='rainbow')
```

# 2. Kmeans 

*Lets try with default params for kmeans clustering*

```python
km = KMeans(n_init='auto')
y = km.fit_predict(scaled)
```

```python
km.cluster_centers_ # 8 centers since n_clusters by default is 8
```

```python
# viz for cluster
sns.scatterplot(x=scaled[:,0],y=scaled[:,1],hue=y,palette='rainbow')
```

```python
# evaluate clusters
silhouette_score(scaled,y)
```

*silhouette score ranges from -1 to 1 where ideal value is 1 since we got 0.23 approx. its not the perfect cluster*

### Elbow Graph

```python
wcss = []

for i in range(1,20): # create n clusters itteratively ranging 1 to 20
    km = KMeans(n_clusters=i) 
    km.fit_predict(scaled)
    wcss.append(km.inertia_) # append the squared distance from closest cluster
```

```python
plt.plot(range(1,20),wcss)
plt.title('Elbow Graph')
plt.axvline(4,linestyle='--',c='r')
```

*for N_cluster = 4 we get ideal score*


### Final Kmeans model

```python
fkmean = KMeans(n_clusters=4,n_init='auto')
ypred = fkmean.fit_predict(scaled)
```

```python
silhouette_score(scaled,ypred)
```

```python
# viz
sns.scatterplot(x=scaled[:,0],y=scaled[:,1],hue=ypred,palette='rainbow')
plt.scatter(fkmean.cluster_centers_[:,0],fkmean.cluster_centers_[:,1],color = 'yellow',s=100) # cluster center
```

### Evaluation

```python
model = SilhouetteVisualizer(KMeans(n_clusters=4))
vis = model.fit(scaled)
vis.show()
```

```python
df['Cluster(kmean)'] = ypred
```

```python
df.head()
```

# 3. DBSCAN

```python
db = DBSCAN(eps=0.5,min_samples=3) # eps = radius min_sample = minimum sample that should be close to core point
y = db.fit_predict(scaled)
```

```python
y
```

*As observed not ideal clustering since every point is considered as noise (i.e -1)*

### Tuning

```python
neigh = NearestNeighbors(n_neighbors=3).fit(scaled)
d,i = neigh.kneighbors(scaled)
```

```python
pd.DataFrame(d)
```

```python
dist = np.sort(d[:,1])
```

```python
plt.plot(dist)
plt.xlabel('Distance')
plt.ylabel('eps')
plt.axhline(1.31,linestyle='--',c='r')
plt.show()
```

```python
dbs = DBSCAN(eps=1.31,min_samples=8)
yp = dbs.fit_predict(scaled)
yp
```

```python
silhouette_score(scaled,yp)
```

```python
# viz
sns.scatterplot(x=scaled[:,0],y=scaled[:,1],hue=yp,palette='rainbow')
```

```python
df['Cluster(DBSCAN)'] = yp
```
