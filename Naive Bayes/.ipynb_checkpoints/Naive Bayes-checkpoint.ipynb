```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from sklearn.metrics import classification_report , confusion_matrix
plt.rcParams['figure.figsize'] = (15,10)
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB , BernoulliNB , GaussianNB
from sklearn.preprocessing import OrdinalEncoder , LabelEncoder
```

# Data

```python
df = pd.read_csv('./SalaryData_Train.csv')
df
```

# EDA

## check for null and duplicate

```python
df.isna().any()
```

```python
df[df.duplicated()]
```

```python
df.drop(index=df[df.duplicated()].index,inplace=True)
```

```python
df.columns
```

```python
X = df.drop('Salary',axis=1)
Y = df['Salary']
```

```python
oe = OrdinalEncoder()
lb = LabelEncoder()
X[['workclass', 'education', 'maritalstatus',
       'occupation', 'relationship', 'race', 'sex', 'native']] = oe.fit_transform(X[['workclass', 'education', 'maritalstatus',
       'occupation', 'relationship', 'race', 'sex', 'native']])
Y = lb.fit_transform(Y)
```

# viz

```python
sns.countplot(data=df,x='Salary',width=0.2)
```

```python
crosstab = pd.crosstab(index=df['maritalstatus'],columns=df['Salary'])
crosstab.plot.bar()
```

```python
crosstab = pd.crosstab(index=df['workclass'],columns=df['Salary'])
crosstab.plot.bar()
```

```python
crosstab = pd.crosstab(index=df['education'],columns=df['Salary'])
crosstab.plot.bar()
```

```python
crosstab = pd.crosstab(index=df['relationship'],columns=df['Salary'])
crosstab.plot.bar()
```

```python
crosstab = pd.crosstab(index=df['sex'],columns=df['Salary'])
crosstab.plot.bar()
```

```python
crosstab = pd.crosstab(index=df['race'],columns=df['Salary'])
crosstab.plot.bar()
```

# model building

```python
xtrain,xval , ytrain,yval = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
nb = MultinomialNB()
nb.fit(xtrain,ytrain)
```

```python
print(f"Train Accuracy : {round(nb.score(xtrain,ytrain),2)} \nTest Accuracy : {round(nb.score(xval,yval),2)}")
```

```python
ypred = nb.predict(xval)
```

```python
print(classification_report(yval,ypred))
```

```python
sns.heatmap(confusion_matrix(yval,ypred),annot=True,fmt=".0f")
plt.ylabel('True')
plt.xlabel('Predicted')
plt.show()
```

## gaussian nb

```python
gnb = GaussianNB()
gnb.fit(xtrain,ytrain)
```

```python
print(f"Train Accuracy : {round(gnb.score(xtrain,ytrain),2)} \nTest Accuracy : {round(gnb.score(xval,yval),2)}")
```
