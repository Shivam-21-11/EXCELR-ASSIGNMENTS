# Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (10,10)
plt.rcParams['figure.dpi'] = 500
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.ensemble import RandomForestRegressor 
from sklearn import tree
from sklearn.metrics import classification_report , confusion_matrix
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (10,5)
```

# Data

```python
df = pd.read_csv('./Company_Data.csv')
df
```

# Data Preprocessing

```python
df=pd.get_dummies(df,columns=['ShelveLoc','Urban','US'], drop_first=True)
```

# Random Forest

```python
# split the feature and target
X = df.drop('Sales',axis=1)
Y = df['Sales']
```

```python
# split train test 
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.1,random_state=42)
```

```python
rf_reg = RandomForestRegressor().fit(xtrain,ytrain)
```

```python
print('Training accuracy: ', rf_reg.score(xtrain,ytrain))
print('Testing Accuracy: ', rf_reg.score(xtest,ytest))
```

# Hyperparameter

```python
params = {'n_estimators':[100,150,300,200,500],
    'criterion' : ['squared_error', 'absolute_error','friedman_mse'],
    'max_depth' : [10,15,25,35],
}
grid = GridSearchCV(RandomForestRegressor(),params,verbose=1).fit(xtrain,ytrain)
```

```python
grid.best_score_ , grid.best_params_ 
```

```python
grid.best_estimator_.score(xtest,ytest)
```

```python
grid.best_estimator_.score(xtrain,ytrain)
```

```python
## Final model
rf_reg = RandomForestRegressor(n_estimators=500,criterion='friedman_mse',max_depth=15).fit(xtrain,ytrain)
```

# Feature Importance

```python
rf_reg.feature_importances_
```

```python
feature_df = pd.DataFrame({'Feature':X.columns,
                          'Importance':rf_reg.feature_importances_})
feature_df.sort_values('Importance',ascending=False,inplace=True)
```

```python
feature_df
```

```python
sns.barplot(data=feature_df,y='Feature',x='Importance',palette='rainbow')
```
